Find the determinant of a matrix using scipy


```python
import scipy.linalg as s

matrix=np.random.randn(4,4)
print("Matrix:\n",matrix)
print("Determinant:",s.det(matrix))
```

Find the eigen value and eigen vector of a matrix using scipy

```python
import scipy.linalg as s

matrix=np.random.randn(3,3)
eigenvalue,eigenvect=s.eig(matrix)
print("Matrix:\n",matrix)
print("Eigenvalue:\n",eigenvalue)
print("Eigenvector:\n",eigenvect)
```

**Feature Normalization: Min-max normalization **



```python
import pandas as pd
import numpy as np

data=[[101,102,103,104,105],[88,92,95,93,np.nan],[84,89,87,86,80],[74,65,87,93,94]]
indices=['StudentID','Math','Science','English']
df=pd.DataFrame(data,index=indices)
print("Data:\n",df)
```

Statistical measure

```python
print("\n--- Statistical Measures ---")
print("Mean:\n", df.mean(numeric_only=True))
print("Median:\n", df.median(numeric_only=True))
print("Mode:\n", df.mode(numeric_only=True).iloc[0])
print("Variance:\n", df.var(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))
```

***Data Preprocessing***

```python

df = df.T
df.columns.name = None
df.reset_index(drop=True, inplace=True)
print("Transformed Data:\n", df)
```

```python
print("\n--- Filtered Students Scoring Above 90 in Science ---")
filtered = df[df['Science'] > 90]
print(filtered)
```

```python
extra_data = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 105],
    'Attendance (%)': [96, 88, 92, 85, 98]
})
merged = pd.merge(df, extra_data, on='StudentID')
print(merged)
```

```python

print("\n Handling Missing Values ")
df['Math'].fillna(df['Math'].mean(), inplace=True)
print("After:\n", df)
```

```python
from sklearn.preprocessing import MinMaxScaler
print("\n--- Min-Max Normalization ---")
scaler = MinMaxScaler()
numerical_cols = ['Math', 'Science', 'English']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
print(df)

```
